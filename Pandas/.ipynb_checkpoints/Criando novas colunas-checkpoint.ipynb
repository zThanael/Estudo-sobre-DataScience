```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/alugel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
dados['Valor m2'] = (dados['Valor'] / dados['Area']).round(2)
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(10)
```

```python
casa = ['Casa','Casa de Condomínio','Casa de Vila']
# Apply() aplica uma função a cada registro
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados.head(10)
```

### Excluindo colunas

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado','Valor m2','Valor Bruto','Valor Bruto m2']])
```

```python
dados_aux.head(4)
```

```python
del dados_aux['Valor Bruto'] # Maneiras de deletar colunas
```

```python
dados_aux.pop('Valor Bruto m2') #pop
```

```python
dados.drop(['Valor Bruto','Valor Bruto m2'], axis = 1, inplace = True)
```

```python
dados
```

```python
dados.to_csv('dados/alugel_residencial.csv', sep=';', index = False)
```
