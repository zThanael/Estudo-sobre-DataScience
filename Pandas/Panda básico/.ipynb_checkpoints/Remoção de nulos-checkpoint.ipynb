```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/alugel_residencial.csv',sep=';')
```

```python
dados
```

#### Tratamento de dados faltantes

```python
dados.isnull()
```

```python
dados.notnull()
```

```python
dados.info()
```

```python
dados[dados['Valor'].isnull()] #dados['Valor'].isnull() mesmo que selecao utilizado antes
```

```python
a = dados.shape[0]
dados.dropna(subset = ['Valor'], inplace=True) #dropna elimina registros nulos, com base na coluna passada no subset = []
b = dados.shape[0]
a-b 
```

#### Continuação

```python
dados[dados['Condominio'].isnull()].shape
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
a = dados.shape[0]
dados = dados[~selecao] #Inverte
b = dados.shape[0]
a-b 
```

#### Trocar os NULL por 0

```python
dados = dados.fillna({'Condominio': 0,'IPTU':0}) 
#Podemos passar um dict, com as colunas e os valores que desejamos trocar para cada coluna
#Também podemos so passar um valor, dados.fillna(0) onde todos os nulos serão trocados por 0
```

```python
dados.info()
```

```python
dados.to_csv('dados/alugel_residencial.csv', sep = ';', index = False)
```

```python
#dados.fillna(method = 'ffill') coloca o valor anterior no local do nulo
#dados.fillna(method = 'bfill') o inverso do ffill, ou seja, coloca o valor depois do null no seu local
#dados.fillna(s.mean()) coloca no local do null a media dos valores não nullos.
#dados.fillna(method = 'ffill',limit=1) limita o tanto de casas nullo que ele vai passar
```
