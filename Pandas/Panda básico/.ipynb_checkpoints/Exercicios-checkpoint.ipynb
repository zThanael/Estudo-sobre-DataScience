```python
import pandas as pd
```

```python
def criar_filtro(df,coluna,lista_criterios):
    selecao = df[coluna].isin(lista_criterios) ## isin() mesmo que dados[coluna] == 'valor'
    return df[selecao]
```

```python
def reiniciar_index(df):
    df.index = range(df.shape[0])
    return df
```

## Exercicio 1
- Selecione somente os imóveis classificados com tipo 'Apartamento'

```python
dados = pd.read_csv('dados/aluguel.csv',sep=';')
```

```python
apartamento = criar_filtro(dados,'Tipo',['Apartamento'])
apartamento = reiniciar_index(apartamento)
```

```python
apartamento
```

## Exercicio 2

```python
residencial = criar_filtro(dados,'Tipo',['Casa','Casa de Condomínio','Casa de Vila'])
```

```python
residencial = reiniciar_index(residencial)
```
